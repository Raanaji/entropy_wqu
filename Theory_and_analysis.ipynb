<h3><center>Brief about VaR and CVaR</center></h3>

VaR in simple terms is a maximum potential loss pertinent to the portfolio under a given confidence level. Technically it is the  measure that tells how much  one can lose under a given  confidence  level  (95%  and  99%)  over  a  given  time  horizon.  This  can  further  be explained with an example of a portfolio of USD 1 million having a 10-day VaR at 95% equal to USD 10000, which means the portfolio cannot lose more than USD 10000 over a 
period of 10 days with a confidence level of 95%. VaR assumes asset returns follow normal distribution. 

There are mainly three types of VaR analysis viz; parametric VaR, Historical 
simulation based VaR and Monte  Carlo simulation based VaR. The main constituents of VaR analysis are portfolio mean, standard deviation and alpha (confidence level). As VaR answers the maximum loss we can incur under a certain confidence level then there arises another question like ‚Äúwhat will be the loss if actual loss exceeds the number obtained from VaR analysis‚Äù. Then CVaR (conditional VaR) comes into picture answering such questions 
by taking the average of losses exceeding the VaR number in analysis. CVaR is also called expected shortfall and is used as an extension to the VaR.

<h3><center>Minimum Variance (Haugen, Baker 1991)</center></h3>

Maximizing returns for minimum while concentrating exclusively on minimizing volatility. 

This is financial application of a basic mathematical optimization problem where the aim is to minimize the variance of the equation representing the portfolio. 


$$\text{Min:}\sigma^{2}_{p}=w'\Sigma\;w\;\;\;\;\;(1)$$
$$\text{s.t.}\;w'l=1$$

where the covariance matrix of the portfolio assets is denoted as $N \times N$ and $\Sigma$ represents the set consisting of prospective investments, $w$ is a $N\times 1$ vector which includes the weight of each asset, finally $l$ is a $N\times 1$ vector consisting of ones. 

This optimization problem has the solution of the form

$$w_{MV}=\frac{\Sigma^{-1}l}{l'\Sigma^{-1}l}\;\;\;\;\;(2)$$

with $\Sigma^{-1}$ representing an inverse matrix such that

$$\Sigma^{-1}=\text{Diag}(1/\sigma^{2})-\frac{(\beta/\sigma^{2})(\beta/\sigma^{2})'}{\frac{1}{\sigma^{2}_{M}}+(\beta/\sigma^{2})'\beta}$$

De silva et. al.(2011,2013)[ref both papers w/ numbers] have shown that the returns on a given $i^{\text{th}}$ can be modelled using the following CAPM equation as 

$$r_i=\alpha_i+\beta_i r_M+Œµ_i\;\;\;\;\;(3)$$


where $\sigma^{2}_{Œµ,i}$ is the idiosyncratic random variable which is not correlated with all other factors within the above (eq3) CAPM model, the error term $Œµ_i$ has mean of zero. 

This gives us an $N\times N$ covariance matrix of the form

$$\pmb{\Sigma}=\pmb{\beta}\pmb{\beta}'\sigma_{M}+\text{Diag}(\pmb{\sigma^{2}_{Œµ}})\;\;\;\;\;(4)$$

where the variance within the CAPM model is denoted by $\sigma_M$, the $N\times 1$ vector of $\beta_{i}$ is denoted by $\pmb{\beta}$, $\text{Diag}(\sigma^{2}_{Œµ})$ being the diagonal matrix consisting of $\sigma_{Œµ,i}$ with null values populated in all places except the diagonal coordinates.

To find the weights for our portfolio allocation, we solve equation(2) and equation(4) whose computational solution is

$$w_{i}=\frac{\sigma^{2}_{MV}}{\sigma^{2}_{Œµ,i}}\bigg(1-\frac{\beta_{i}}{\beta_{LS}}\bigg)\;\;\;\;\;(5)$$

such that the beta of the market (ex ante) is given by $\beta_{i}$. The beta threshold **Long-Short** $\beta_{LS}$ is a determining variable (for the weights) in an unconstrainted portfolio optimization, and is represented by the implicit solution

$$\beta_{LS}=\frac{\frac{1}{\sigma^{2}_{M}}+\Sigma\frac{\beta^{2}_{i}}{\sigma^{2}_{Œµ,i}}}{\Sigma\frac{\beta^{2}_{i}}{\sigma^{2}_{Œµ,i}}}.\;\;\;\;\;(6)$$

Thus, from our above solution for individual weights equation(5) we have two cases:

- Those securities which have $\beta_i<\beta_{LS}$ will have positive weights
- Those securities which have $\beta_i>\beta_{LS}$ will have negative weights

For the **long only** case we will discard the second case and only consider the first. In doing so we have a conditional expression as follows;

$$w_i=
\left\{ 
  \begin{array}{ c l }
    \frac{\sigma^{2}_{LMV}}{\sigma^{2}_{Œµ,i}}\bigg(1-\frac{\beta_{i}}{\beta_{L}}\bigg) & \quad \textrm{if } \beta_i<\beta_{L} \\
    0,                 & \quad \textrm{otherwise}
  \end{array}
\right.\;\;\;\;\;(7)$$

here the long only MV-portfolio's variance (ex ante) is denoted by $\sigma_{LMV}$. The threshold beta for the long only portfolio is denoted by $\beta_L$ which has the following expression;

$$\beta_{LS}=\frac{\frac{1}{\sigma^{2}_{M}}+\Sigma\beta_i<\beta_{L}\frac{\beta^{2}_{i}}{\sigma^{2}_{Œµ,i}}}{\Sigma\beta_i<\beta_{L}\frac{\beta^{2}_{i}}{\sigma^{2}_{Œµ,i}}}.\;\;\;\;\;(8)$$


Now let us evaluate heuristically equation(7). Let us say that the market usually has a high volatility which is denoted by the idiosyncratic variable $\sigma^{2}_{Œµ,i}$. Also, let us assume the investor would like to aim for higher returns (which are complemented by higher market volatility) which points to high market beta. As we can see, this invariably push the equation(7) to zero. Thus, we can straightaway conclude that many of the securities in our portfolio will either have smaller weights or none at all.

The minimum variance method lets us disqualify putting efforts into securities which in an overall sense are not investable for our portfolio. This helps us concentrate on exploring other financial metrics for the selected securities.

<h3><center>Maximum Diversification (Choueifaty, Coignard 2008)</center></h3>


To summarize the maximum diversification method introducted by Choueifaty et. al. (2008), investments in risk-efficient markets will produce returns in proportion to their total risk. 

To measure how diverified a portfolio is, we take a simple ratio with the  weighted average assets volatility in the numerator to portfolio volatility in denominator. This ratio is maximized as a part of the basic optimization problem. The maximum diversification method takes (all) asset's marginal contribution (to the portfolio) and equalizes it for a change in asset's weight.

We defined the diversification ratio (being maximized) as follows;

$$\text{Max:}D(\pmb{w})=\frac{\pmb{w}'\pmb{\sigma}}{\sqrt{\pmb{w}'\pmb{\Sigma}}\pmb{w}},\;\;\;\;\;(9)$$

where the asset's volatility is denoted by a $N \times 1$ vector $\pmb{\sigma}$, and the diversification ratio is denoted by the variable $D(\pmb{w})$.

**Assumption**: Let $R$ be the returns on the assets such that $R(w)$ is the returns as a function of the weights $w$. Then we can assume that the risk is directly proportional to the expected additional returns on the asset. We mathematically formulate as follows;

$$E[R(\pmb{w})]=k\pmb{w'}\pmb{\sigma},,\;\;\;\;\;(10)$$

where $k$ being a proportionality constant. Substituting eq10 in eq9 we get,

$$\text{Max:}D(\pmb{P})=\frac{E[R(\pmb{w})]}{\sqrt{\pmb{w}'\pmb{\Sigma}}\pmb{w}},\;\;\;\;\;(11)$$

where $D(P)=\frac{\pmb{w}}{k}$.

Thus, maximizing eq11 will be will yield us the Sharpe ratio of our portfolio.

An advantage of using maximum diversification method is that the weights tend to be less concentrated in our portfolio (compared to minimum variance), however the method suffers from the same drawbacks as the minimum variance.

Primary principle is all assets have similar returns and risks but heterogeneous correlations. 

Maximum decorrelation is fairly simple, the portfolio is optimal if all the assets somehow have equal expected return and equal volatilities, bar heterogenous correlations.

$$\text{s.t.}\;\pmb{1}^{T}\pmb{w}=1$$

Thus, we rely only on the correlation as a measure of variation in this method.


<h3><center>Equal Risk Contribution (a.k.a Risk Parity Contribution) (Mailard, Roncalili, Teiletche 2008)</center></h3>


As the name suggests the method is about equalizing the risk from each asset in the portfolio. For a set of $n$ risky assets let $w=(w_1,...,w_n)$ represent a set of weights, then 

$$c_{i}(w) = \frac{(\Sigma\pmb{w})_{i}}{\sqrt{\pmb{w}'\Sigma\pmb{w}}} = \frac{\partial\sigma(w)}{\partial w_{i}} = \frac{\partial\sqrt{\pmb{w}'\Sigma\pmb{w}}}{\partial w_{i}}\;\;\;\;\;(12)$$


is the marginal volatility contribution of an asset $i$ which represents the change in volatility which is a consequence of minor changes caused within the portfolio by the changes of the weights of an asset. Here the $(\Sigma\pmb(w))$ is a $N\times1$ dimensional vector.

We can decompose the portfolio risk by using the following expression;

$$\sigma(\pmb{w})=\sum^{n}_{i=1}w_{i}c_{i}(\pmb{w})=\pmb{w}'\frac{\Sigma \pmb{w}}{\sqrt{\pmb{w}'\Sigma \pmb{w}}}=\sqrt{\pmb{w}'\Sigma \pmb{w}},\;\;\;\;\;(13)$$

where the $i^{\text{th}}$ assets' total risk contribution is denoted by $$\sigma_{i}(\pmb{w})=w_{i}\times c_{i}(\pmb{w}).$$

By solving the following optimization problem the weights can be found as follows;

$$\text{Min:}\sum^{n}_{i=1}(\sigma(\pmb{w})/n-w_{i}c_{i}(\pmb{w}))^{2},\;\;\;\;\;(14)$$
$$\text{s.t.}\sum^{n}_{i=1}w_{i}=1$$


This method of equal risk contribution is a compromise between other methods wherein the goal of optimization is equal volatility for all assets.

<h3><center>Maximum Entropy (Geman et al. 2015)</center></h3>

Some introduction to shannon entropy

**Shannon Entropy**

Say we have a discrete random variable $X$ with a set of possible outcome contained in the set ${x_1,...,x_n}$, with each outcome having a probability $P(x_1),...,P(x_n)$. Then the Shannon entropy for the random variable $X$ is represented by the following expression,

$$H(X)=-\sum^{n}_{i=1} P(x_i) \log P(x_i)$$

where the expression $H(X)$ gives the level of entropy (on average) in the random variables' outcomes.

We will be using a modified form of Shannon entropy where the probability outcome will be replaced by the weight $w_i$. We have borrowed this idea from J, Lee, Chiou (2014). The modified Shannon entropy is then of the form, 

$$H=-\sum^{n}_{i=1}w_{i}\ln w_{i},\;\;\sum^{n}_{i=1}w_{i}=1,\;\;1,2,...,n\;\;\;\;\;(a)$$

where there are $n$ investable assets. Since $\sum^{n}_{i=1}w_{i}=1$ $w_i = \frac{1}{n}$, we can thus see that this is an optimization problem with a constraint. Our aim is to maximize the entropy function $H$, meaning our observations (data) will relay more information. This also includes cases of extremity where the wight will either completely be one (100%) or have no relevant value at all (zero). 

There have been additional modifications to the mean-variance model to comply with the construction using Shannon entropy as the focal point. Ke and Zhang(2008) have proposed the following modifications to the mean-variance procedure;

$$\text{Min:}\;w^T \Omega w+\sum^{n}_{i=1}w_i \ln w_i\;\;\;\;\;(b)$$
$$\text{s.t.} \sum^{n}_{i=1}r_i w_i = E,$$
$$\sum^{n}_{i=1} w_i=1$$
$$w:(w_1,...,w_n)^T$$,

where the covariance matrix is denoted by $\Omega$, the returns (on average) are denoted by $r_i$.

Zheng et al. (2009) have proposed to replace the variance within the model with entropy by using the following method;

$$\text{Max}-\sum^{n}_{i=1}w_i \ln w_i\;\;\;\;\;(c)$$
$$\text{s.t.}\sum^{n}_{i=1}r_i w_i = E$$
$$\sum^{n}_{i=1}\sum^{m}_{t=1}w_i(r_{it} - \bar{r_i})^{2}\leq \text{Var},$$
$$\sum^{n}_{i=1} w_i=1$$

where the variance is $\text{Var}$ and the return of the asset at time $t$ is $r_{it}$.

For further discourse on the optimization of the lagrangian function using the above equations, please refer to Jiang et. al. (2008) for a more copious discussion, wherein the solution ultimately leads to the weights which are of the form;

$$L=-\sum^{n}_{i=1}w_i \ln w_i+\gamma_1 \bigg(\sum^{n}_{i=1}r_i w_i - E\bigg)+\gamma_2 \bigg(\sum^{n}_{i=1}\sum^{m}_{t=1}w_i(r_{it} - \bar{r_i})^{2}- \text{Var}\bigg)+\gamma\bigg(\sum^{n}_{i=1} w_i-1\bigg)\;\;\;\;\;(d)$$

which yields the solution

$$w_i = \frac{e^{[\gamma_{1} \bar{r_i}+\gamma_{2} \sum^{m}_{t=1} (r_{it} - \bar{r_i})^{2}]}}{\sum^{n}_{i=1} e^{[\gamma_{1} \bar{r_i}+\gamma_{2} \sum^{m}_{t=1} (r_{it} - \bar{r_i})^{2}]} }.\;\;\;\;\;(e)$$

we shall now discuss the proof in a simplified manner to retain the financial core of our exercise and not wither away in the mathematical depths of the subject.


<h4><center>Proof of Maximum Entropy</center></h4>

We shall take the above equations (refer equations and rephrase this statement) and solve the following lagrangian problem.

**Proof:** We start by 

$$\text{Max:}-\sum^{n}_{i=1}w_{i}\ln w_{i}$$
$$\text{s.t.}\sum^{n}_{i=1}w_{i}=1$$

$$\text{Min:}\;w^T \Omega w+\sum^{n}_{i=1}w_i$$

thus we have a lagrange equation of the form;

$$L(w, \lambda)=-\sum^{n}_{i=1}w_i \ln w_i +(\lambda + 1)\bigg(\sum^{n}_{i=1} w_i-1\bigg)$$

Imposing the first order condition;



$$\text{First Order Condition}=
\left\{ 
  \begin{array}{ c l }
    \frac{\partial L}{\partial w_i}=(1+\ln w_i)+(\lambda +1)=0 & \quad \textrm{  }  \\
    \frac{\partial L}{\partial \lambda}=\sum^{n}_{i=1}w_{i}-1=0,                 & \quad 
  \end{array}
\right.$$

solving the first order condition on the lagrangian equation gives us 

$$w_i=e^{\lambda}$$
$$ne^{\lambda}=1$$


which gives us $w_i = \frac{1}{n}$, same result as that from equation(a), meaning it is possible to maximize the entropic function and get the resultant solution as weights.
